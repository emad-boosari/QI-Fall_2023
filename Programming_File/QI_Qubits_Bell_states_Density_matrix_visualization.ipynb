## Further Study about the single Qubit gates

---



https://learn.qiskit.org/course/ch-states/single-qubit-gates

```python
q = QuantumRegister(1)
```

```python
c = ClassicalRegister(1)
```

```python
qc = QuantumCircuit(q,c)
qc.x(0)
qc.h(0)
```

```python
qc.draw('mpl', initial_state = True)
```

```python
def qplot(x):
  return x.draw('mpl', style = 'iqx', initial_state = True)
```

```python
qplot(qc)
```

```python
ψ1 = Statevector(qc)
```

```python
ψ1.draw('latex')
```

```python
ρ1 = DensityMatrix(qc)
ρ1.draw('latex',prefix = '\\rho_1 = ')
```

```python
plot_state_city(qc)
```

```python
plot_state_hinton(qc)
```

```python
plot_state_paulivec(qc)
```

```python
plot_bloch_multivector(qc)
```

```python
qc.measure(q[0],c[0])
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots = 2000)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```

## Bell State

```python
bell_circuit0 = QuantumCircuit(2,2)
bell_circuit0.h(0)
bell_circuit0.cx(0,1)

display(qplot(bell_circuit0), Statevector(bell_circuit0).draw('latex'))
```

```python
bell_circuit0.measure(0,0)
bell_circuit0.measure(1,1)
qplot(bell_circuit0)
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(bell_circuit0, backend, shots = 2000)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```

## 3rd Bell-State

```python
bell_circuit2 = QuantumCircuit(2,2)

bell_circuit2.h(0)
bell_circuit2.cx(0,1)
bell_circuit2.x(0)

display(qplot(bell_circuit2), Statevector(bell_circuit2).draw('latex'))
```

### 4th Bell state

```python
bell_circuit4 = QuantumCircuit(2,2)
bell_circuit4.x(0)
bell_circuit4.h(0)
bell_circuit4.cx(0,1)
bell_circuit4.x(0)

display(qplot(bell_circuit4), Statevector(bell_circuit4).draw('latex'))
```
